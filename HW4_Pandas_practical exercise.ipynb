## Series & Data frame 
### series - 인덱스(index)가 있는 1차원 배열 

```python
s = pd.Series([1, 3, 5, 7, 9])
print(s)
print(s[2]) # 인덱싱 가능!
```

```python
# dictionary를 이용
dic = {'a':1, 'b':3, 'c':5}
s = pd.Series(dic)
print(s)
print(s['b']) 
```

```python
# Series의 두 속성 - values, index
print(s.values)
print(s.index)
```

```python
s = pd.Series([1,3,5,7,9], index=['LGT','DOO','HWE','LOT','SKW']) # index를 부여! 
print(s[:'LOT']) # 'LOT' 인덱스까지만 추출!
```

```python
#데이터 프레임 만들기 - 행열로 이루어진 2차원의 표!라고 생각하면 좋음
score={'학번':[201621451,201621223,201621233],
       '이름':['한우주','박용택','정성훈'],
       'Quiz':[20,20,10],
       '중간고사':[20,20,15]}

score_df= pd.DataFrame(score)
score_df
```

### loc, iloc을 이용한 인덱싱과 슬라이싱

#### iloc - 저장된 순서에 따라 0부터 정수로 indexing, slicing 가능 / '위치'가 중요
#### loc - 데이터 프레임에 정의된 index에 대해 언덱싱 작동 / '라벨'이 중요

```python
df=df.reindex([1,4,6,2,3,5,9,8,0,7,11,10]) # 인덱스를 새로 부여
df
```

```python
df.loc[:6] # 인덱싱 값을 반영하여 결과를 반환 
```

```python
df.iloc[:6] # 순서(0~5)까지를 반환 (인텍싱한 값을 반영하지X)
```

```python
df1=df.sort_index() #인덱스 값에 따라서 정렬
df1
```

```python
df1.loc[:6] #인덱스 값이 6인 것을 포함해서 반환
```

```python
df1.iloc[:6] #순서가 6번쨰 전 (5번쨰)까지 반환
```

```python
df2=df.sort_index(ascending=False) #내림차순으로 정렬
df2
```

```python
df2.iloc[:, 2] #세번째 열 (Score) 를 추출
```

```python
df2.loc[:, ['Name', 'Age']] # Name과 Age 열 추출
```

```python
df2.iloc[:, :2] #두번째 열까지 모든 값 출력
```

```python
df2.loc[:, :'Age'] #'Age'열 까지 모든 값 출력
```

### 그룹화 - grouping

#### groupby()를 이용해서 특정 속성별로 그룹화할 수 있음 
#### df['Wscore'].mean() - 모든 행에 있는 Wscore의 평균을 계산
#### groupby("Wteam)을 이용 - Wteam 값이 같은 행들에 대한 Wscore의 평균을 구할 수 있음!

```python
#진 팀의 Wscore의 평균을 구함
df.groupby('Wteam')['Wscore'].mean().head() 
```

```python
#이긴 팀 별로 그룹화 하여, home, away, neutral에서 각각 몇번 이겼는지 구함
df.groupby('Wteam')['Wloc'].value_counts().head(9)
```
